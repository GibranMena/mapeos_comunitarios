```python
# Load Kobo API token and endpoint from .env file
from koboextractor import KoboExtractor
from dotenv import load_dotenv
import os
import requests
import pandas as pd

load_dotenv()  # take environment variables from .env
KOBO_TOKEN = os.getenv("KOBO_TOKEN")
KOBO_ENDPOINT = os.getenv("KOBO_ENDPOINT", "https://eu.kobotoolbox.org/api/v2")

kobo = KoboExtractor(
    token=KOBO_TOKEN,
    endpoint=KOBO_ENDPOINT
)

# Get your forms and data
assets = kobo.list_assets()
asset_uid = assets['results'][0]['uid']  # Use the first asset, or select as needed
data = kobo.get_data(asset_uid)
df = pd.DataFrame(data['results'])
```

```python
# Retrieve data from a specified asset UID and load into a DataFrame

caminata_uid = "aJC29VzyyWUQvBo2uQMXWG"  
specified_data = kobo.get_data(caminata_uid)
caminata_df = pd.DataFrame(specified_data['results'])
caminata_df.head()
```

```python
def get_first_image_url(attachments):
    if isinstance(attachments, list) and attachments:
        for att in attachments:
            if att.get('mimetype', '').startswith('image/'):
                return att.get('download_url')
    return None

# Apply to caminata_df, not resistencias_df
caminata_df['popup_image_url'] = caminata_df['_attachments'].apply(get_first_image_url)
```

```python
import folium
from folium.features import CustomIcon

# Calculate valid geolocations
valid_geos = caminata_df['_geolocation'].dropna().apply(
    lambda x: x if isinstance(x, list) and len(x) == 2 and all(isinstance(i, (float, int)) for i in x) else None
).dropna().tolist()

if valid_geos:
    avg_lat = sum([loc[0] for loc in valid_geos]) / len(valid_geos)
    avg_lon = sum([loc[1] for loc in valid_geos]) / len(valid_geos)
    min_lat = min([loc[0] for loc in valid_geos])
    max_lat = max([loc[0] for loc in valid_geos])
    min_lon = min([loc[1] for loc in valid_geos])
    max_lon = max([loc[1] for loc in valid_geos])
else:
    # Fallback to Berlin if no valid geolocations
    avg_lat = max_lat = min_lat = 52.5200
    avg_lon = max_lon = min_lon = 13.4050

m = folium.Map(
    location=[avg_lat, avg_lon],
    zoom_start=11,  # Initial zoom, will be adjusted by fit_bounds
    tiles='CartoDB positron',
    attr='Map tiles by Stamen Design, CC BY 3.0 ‚Äî Map data ¬© OpenStreetMap contributors'
)

icon_url = 'https://datacritica.org/wp-content/uploads/2025/08/Screenshot-2025-08-27-at-17.36.07.png'

for idx, row in caminata_df.iterrows():
    geo = row.get('_geolocation')
    if not (isinstance(geo, list) and len(geo) == 2 and all(isinstance(x, (float, int)) for x in geo)):
        continue
    lat, lon = geo
    image_url = row.get('popup_image_url')
    image_html = ''
    if image_url:
        image_html = f'<br><img src="{image_url}" width="200">'
    popup_html = f"""
    <b>{row.get('nombre_accion', '')}</b><br>
    Descripci√≥n: {row.get('descripcion_accion', '')}<br>
    Comunidad de inter√©s: {row.get('pueblo', '')}<br>
    Contribuci√≥n de: {row.get('Opcionalmente_escri_e_de_tu_contribuci_n', '')}<br>
    {image_html}
    """
    icon = CustomIcon(icon_url, icon_size=(30, 30))
    folium.Marker(
        location=[lat, lon],
        icon=icon,
        popup=folium.Popup(popup_html, max_width=300),
    ).add_to(m)

# Adjust map to fit all points (zoom to bounds)
if valid_geos and (min_lat != max_lat or min_lon != max_lon):
    m.fit_bounds([[min_lat, min_lon], [max_lat, max_lon]])
```
